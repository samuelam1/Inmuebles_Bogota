A partir de la distribucion de los inmuebles, vamos a relacionar los siguientes tipos de inmueble: Apartamento, Casa y Local

# Precio del m2 por Barrio y hacer el grafico mas adecuado para esta nueva variable

Anteriormente teniamos ya el promedio del area de las ciudades proncipales (Top 10), con el siguiente codigo agruparemos por barrio y vamos a calcular el promedio del precio por metro cuadrado. Dividiendo el promedio del precio en millones entre el promedio del área.

Este gráfico mostrará fácilmente la comparación de los promedios del área y del precio por metro cuadrado de los 10 barrios principales, y permitirá que se puedan comparar fácilmente los valores para cada barrio.

```python
promedio_precio = inmuebles[inmuebles["Barrio"].isin(top10.index)].groupby("Barrio")["Precio_Millon"].mean()

precio_m2=promedio_precio/promedio_area

precio_m2

```

```python
import seaborn as sns 
import matplotlib.pyplot as plt

#Aqui se crea un Daraframe con los promedios del area y del precio por metro cuadrado 

promedios= pd.DataFrame ({
    "Promedio Area":promedio_area,
    "Promedio Precio/m2":precio_m2
})

#Crear el grafico de barras agrupadas 

ax=sns.barplot(data=promedios,x=promedios.index, y="Promedio Area",color="blue")
ax=sns.barplot(data=promedios,x=promedios.index, y="Promedio Precio/m2",color="orange")

#Agregar leyyendas y titulos 

ax.set_ylabel("Promedio")
ax.legend(labels=["Area", "Precio/m2"], loc="upper left")
plt.title("Promedio de area y precio por metro cuadrado de los 10 barrios proncipales")


# Rotar las etiquetas del eje X
ax.set_xticklabels(ax.get_xticklabels(), rotation=90, ha="right")

plt.show()





```
